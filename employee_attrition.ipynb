```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
import pandas as pd
```

```python
categorical_features = x.select_dtypes(include=['object']).columns
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), x.select_dtypes(exclude=['object']).columns),
        ('cat', OneHotEncoder(sparse_output=False, handle_unknown='ignore'), categorical_features)
    ])
```

```python
x_scaled = preprocessor.fit_transform(x)
```

```python
x_scaled = pd.DataFrame(x_scaled, columns=preprocessor.get_feature_names_out())
```

```python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()

```

```python
categorical_features = x.select_dtypes(include=['object']).columns.tolist()  # Get columns with object dtype
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(sparse_output=False, handle_unknown='ignore'), categorical_features)], remainder='passthrough')
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
x_train = ct.fit_transform(x_train)
x_test = ct.transform(x_test)
std_scaler = StandardScaler()
xtrain_scaled = std_scaler.fit_transform(x_train)
xtest_scaled = std_scaler.transform(x_test)
```

```python
xtrain_scaled
xtest_scaled
```

```python
from sklearn.ensemble import RandomForestClassifier
Ramdom_forest_model=RandomForestClassifier()
```

```python
Ramdom_forest_model.fit(xtrain_scaled,y_train)
```

```python
y_pred=Ramdom_forest_model.predict(xtest_scaled)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cn = confusion_matrix(y_test,y_pred)
print(cn)
```

```python
sns.heatmap(cn,annot=True,fmt='d')
```

```python
from sklearn.metrics import accuracy_score
model_accuracy = accuracy_score(y_test,y_pred)
print("accuracy of the model = ",model_accuracy)
```

```python
from sklearn.metrics import precision_score
model_precision = precision_score(y_test, y_pred, pos_label='Yes')
print("precision of the model = ", model_precision)
```

```python
from sklearn.metrics import recall_score
model_recall = recall_score(y_test, y_pred, pos_label='Yes')
print("recall of the model = ", model_recall)
```

```python
from sklearn.metrics import f1_score
model_f1 = f1_score(y_test, y_pred, pos_label='Yes')
print("f1 score of the model = ", model_f1)
#
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
df = pd.read_csv('/content/WA_Fn-UseC_-HR-Employee-Attrition.csv')
X = df.drop('Attrition', axis=1)
score_list = Ramdom_forest_model.feature_importances_
list_of_features = list(X.columns)
feature_importances_dict = dict(zip(list_of_features, Ramdom_forest_model.feature_importances_))
score_list_adjusted = [feature_importances_dict[feature] for feature in list_of_features if feature in feature_importances_dict]
score_df = pd.DataFrame({"Feature": list_of_features, "score": score_list_adjusted})
score_df.sort_values(by="score", ascending=False)
score_df.sort_values(by ="score", ascending=False)
```

```python
y = df['Attrition']
categorical_features = X.select_dtypes(include=['object']).columns
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore') # parse=False for dense output
encoded_data = encoder.fit_transform(X[categorical_features])
encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(categorical_features))
X = X.drop(categorical_features, axis=1)
X = pd.concat([X, encoded_df], axis=1)
Random_forest_model = RandomForestClassifier(n_estimators=100, random_state=42)
Random_forest_model.fit(X, y)
Random_forest_model = RandomForestClassifier(n_estimators=100, random_state=42)
Random_forest_model.fit(X, y)
list_of_features = list(X.columns)
plt.figure(figsize = (30,18))
plt.barh(range(len(list_of_features)),Random_forest_model.feature_importances_)
plt.yticks(range(len(list_of_features)),list_of_features)
plt.xlabel('feature importance')
plt.show()
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(Random_forest_model, X, y, cv=5)
print("Cross-validation scores:", scores)
```

```python
Avg_Model_score = scores.mean()
print('Averagae Model score = ',Avg_Model_score )
```
